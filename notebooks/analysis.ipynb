```python
import json

json_data = []
with open("../com/njcap/scripts/funds_stats/stats.json") as statsFile:
    json_data = statsFile.readline()
    json_data = json.loads(json_data)
    print(json_data)
```

```python
json_data[0]
```

```python
# now we will open a file for writing 
import csv
with open('data_file.csv', 'w') as data_file:
  
    # create the csv writer object 
    csv_writer = csv.writer(data_file) 
    
    csv_json = []
    
    for obj in json_data:
        csv_json_object = {}
        csv_json_object.update(obj["meta"])
        csv_json_object.update(obj["data"])
        csv_json.append(csv_json_object)

    # Counter variable used for writing  
    # headers to the CSV file 
    count = 0

    for fund in csv_json: 
        if count == 0: 

            # Writing headers of CSV file 
            header = fund.keys() 
            csv_writer.writerow(header) 
            count += 1

        # Writing data of CSV file 
        csv_writer.writerow(fund.values())
```
